Task 1: Graph a function of one variable Plot the graph of sin(x) for x in the interval
[0, 10].

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 1000) # Generate x from 0 to 10 with a small step size for a smooth curve

y = np.sin(x) # Calculate sin(x) for each x

# Create the plot
plt.plot(x, y, label='sin(x)', color='b')
plt.title('Graph of sin(x) for x in [0, 10]')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)
plt.show()
```

√ Task 2: Graph a function of two variables Plot the graph of cos( x2 + y2) for (x, y) ∈
[−5, 5] × [−5, 5].

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a grid of x and y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Calculate the corresponding z values using the function f(x, y) = cos(x^2 + y^2)
Z = np.cos(X**2 + Y**2)

fig = plt.figure() #construct a 3d graph
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(X, Y, Z, cmap='viridis') # Plot the surface
fig.colorbar(surf) # Add color bar

# Set axis labels
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

# Set title
ax.set_title('Graph of cos(x^2 + y^2)')
plt.show()
```
