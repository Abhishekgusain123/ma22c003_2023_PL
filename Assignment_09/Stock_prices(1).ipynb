• Find out if a heirarchical indexing (MultiIndex) is being used in this data.

```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import yfinance as yf
import pandas as pd

# Define the ticker list
tickers_list = ['AAPL', 'GOOG', 'TSLA']

# Download current stock prices
data = yf.download(tickers_list, start='2015-01-01')['Adj Close']
print(data.head())
```

```python
# Check if a MultiIndex is being used
# Checks if the index of the DataFrame data is a MultiIndex and stores the result in the is_multi_index variable.
is_multi_index = isinstance(data.index, pd.MultiIndex)

# Print the result based on whether a MultiIndex is being used or not
if is_multi_index:
    print("The data has a MultiIndex.")
else:
    print("The data does not have a MultiIndex.")
```

• Access and plot the closing price of AAPL on all days in the data.

```python
# Calculate cumulative returns and plot
cumulative_returns = ((data.pct_change() + 1).cumprod())
cumulative_returns.plot(figsize=(12, 6))
plt.legend(tickers_list)

# Define the label for the title of the figure
plt.title("Cumulative Returns of Stocks", fontsize=16)

# Define the labels for x-axis and y-axis
plt.ylabel('Cumulative Returns', fontsize=14)
plt.xlabel('Year', fontsize=14)

# Plot the grid lines
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)

plt.show()
```

• Print the closing price of all three stocks yesterday.

```python
# Select yesterday's data and extract the closing prices for all three stocks
yesterday_data = data.iloc[-1]
closing_prices_yesterday = yesterday_data

# Print the closing prices for all three stocks
print("Closing Prices for Yesterday:")
print(closing_prices_yesterday)

# Extract the closing price data for Apple (AAPL)
aapl_closing_price = data['AAPL']

# Plot the closing price of AAPL
plt.figure(figsize=(12, 6))
aapl_closing_price.plot(label='AAPL Close Price')
plt.title('AAPL Closing Price Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.legend()
plt.grid()
plt.show()
```
