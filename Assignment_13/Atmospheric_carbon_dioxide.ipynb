Task: Reconsider the problem of the assignment on rising atmospheric CO2 levels. Using
the data you already downloaded for the assignment, now ft a quadratic curve to the
data by regression. Since the derivative of a quadratic function can be hand-computed,
you would then be able to estimate the rate of change of the regression ft and address the
earlier assignment task: provide a policy-maker with a “yes or no” answer on whether the
rate of increase of CO2 is increasing.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Downloaded CO2 data from the CSV file
df = pd.read_csv('/content/co2_mm_mlo.csv')

# Extract the relevant columns (assuming columns 2 and 3 represent time and CO2 levels)
x = np.array(df.iloc[:, 2])
y = np.array(df.iloc[:, 3])

# Define the quadratic function
def quadratic_function(x, a, b, c):
    return a * x**2 + b * x + c

# Fit the quadratic curve to the data
params, _ = curve_fit(quadratic_function, x, y)

# Extract the coefficients a, b, and c
a, b, c = params

# Calculate the derivative of the quadratic function
derivative = 2 * a * x + b

# Determine if the rate of CO2 increase is accelerating or decelerating
if a > 0:
    result = "Yes, the rate of CO2 increase is increasing."
elif a < 0:
    result = "No, the rate of CO2 increase is decreasing."
else:
    result = "The rate of CO2 increase is stable."

plt.scatter(x, y, label="Data")
plt.plot(x, quadratic_function(x, a, b, c), label="Quadratic Fit")
plt.xlabel("Time")
plt.ylabel("CO2 Levels")
plt.legend()
plt.title(result)
plt.show()

print("Quadratic Coefficients (a, b, c):", a, b, c)
print("Derivative (2ax + b):", derivative)
print("Conclusion:", result)
```
